```python
#To Check if there are continenets that have grown over the years
df.groupby('Continent').mean()
```

```python
# To look up for Oceania
df[df['Continent'].str.contains('Oceania')]
```

```python
df2 = df.groupby('Continent').mean().sort_values(by = '2022 Population', ascending = False)
df2
```

```python
# To transpose the header
df3 = df2.transpose()
df3
```

```python
# Visualizing Average Population
df3.plot()
```

```python
# To specify the various columns you want to visualize
df.columns
```

```python
df2 = df.groupby('Continent')[['2022 Population', '2020 Population', '2015 Population',
       '2010 Population', '2000 Population', '1990 Population',
       '1980 Population', '1970 Population']].mean().sort_values(by = '2022 Population'
       , ascending = False)
df2
```

```python
# Or you can select the columns this way
df.columns
```

```python
df2 = df.groupby('Continent')[df.columns[5:13]].mean().sort_values(by = '2022 Population', ascending = False)
df2
```

```python
# We can now transpose df2
df3 = df2.transpose()
df3
```

```python
# Lets plot the df3
df3.plot()
```

```python
# To order the years from oldest to recent for proper visualization
df2 = df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population',
       '2000 Population', '2010 Population', '2015 Population',
       '2020 Population', '2022 Population']].mean().sort_values(by = '2022 Population'
       , ascending = False)
df2
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot()
```
